```python
#Преобразуем колонку kepler_name и заменим пустые значения на среднее

def get_num(kep_name):
    if type(kep_name) == str:
        try:
            return float(kep_name.split(" ")[0].split("-")[1])
        except ValueError:
            return 0
    else:
        return float(kep_name)
    
kepler_df['kepler_num'] = kepler_df['kepler_name'].apply(get_num)
kepler_df['kepler_num'].fillna(int(kepler_df['kepler_num'].mean())).astype(int)
```

```python
#2.9. создание нового поля вычисленного на основе значений других полей:
# 2.9.1 через выражение на базе имеющихся колонок,
kepler_df['koi_slogg_err_dist'] = kepler_df['koi_slogg_err1'] - kepler_df['koi_slogg_err2']
kepler_df['koi_slogg_err_dist']
```

```python
# 2.9.2 через DataFrame.apply
insol_df = kepler_df['koi_insol']
insol_df_median = insol_df.median()
#insol_df_median
insol_df.apply(lambda x: x > insol_df_median).rename('More Than Median Insolation Flux')
```

```python
# 2.9.3 через Series.apply
# 1 sr = 696 000 km
kepler_df['koi_srad_km'] = kepler_df['koi_srad'].apply(lambda x: x * 696000)
kepler_df['koi_srad_km']
```

4 Решейпинг данных 1Dto2D с помощью .pivot (можно подать на вход результаты агрегатов, полученных ранее через .groupby (сгруппировать по двум полям), либо прекрасно заходит сюда данные из SQL сгруппированные предварительно по 2-м полям). Какую информацию о данных вы получили, выполнив данную операцию? Запишите ответ в комментариях.


```python
teq_df = kepler_df[['koi_disposition', 'koi_teq', 'koi_insol']].copy()
teq_df['teq_less_than_400'] = teq_df['koi_teq'] < 400
teq_df = teq_df.drop('koi_teq', axis=1)
```

```python
teq_df.groupby(['koi_disposition', 'teq_less_than_400'], as_index=False).mean().pivot(index='koi_disposition', columns='teq_less_than_400')
```

Как можно увидеть, планетарная равновесная температура зависит от потока инсоляции: там, где она превышает 400К, поток имеет интенсивность в сотни и тысячи раз выше.
